Q1:-Consider the given list as [2, 7, 4, 1, 3, 6]. Write a program to count pairs of elements with
sum equal to 10.

```python
# Definining the list
List = [2, 7, 4, 1, 3, 6]

# Initializing count variable
count = 0

# Checking for all the pairs
for i in range(len(List)):
    for j in range(i + 1, len(List)):  # i kept j starting from i+1 to avoid repeating pairs
        if List[i] + List[j] == 10:  # Checking if sum is 10
            count += 1  # Increase the count variable if pair is found
            print(f"Pair found: ({List[i]}, {List[j]})")  # Printing the pair values if found
        else:
            print(f"Pair ({List[i]}, {List[j]}) does NOT sum to 10")  # else it will return the pairs that don't sum to 10

# Printing the total pairs
if count > 0:
    print(f"Total pairs of elements with sum equal to 10 are: {count}")
else:
    print("No pairs found with sum equal to 10")
```

Q2:- Write a program that takes a list of real numbers as input and returns the range (difference between minimum and maximum) of the list. Check for list being less than 3 elements in which case return an error message (Ex: “Range determination not possible”). Given a list [5,3,8,1,0,4], the range is 8 (8-0).

```python
def calculate_range(numbers):
    # Checks if the list has lesser than 3 elements
    if len(numbers) < 3:
        return "Range determination not possible"

    # Finds the minimum and maximum
    min_val = min(numbers)
    max_val = max(numbers)

    # Calculate range
    return max_val - min_val


# Main program
if __name__ == "__main__":

    real_numbers = input("Enter the numbers with spaces in between: ")
    # it Converts the input string into list of integers
    data = list(map(int, real_numbers.split()))

    # Calls the function and prints result
    result = calculate_range(data)
    print("Range:", result)
```

Q3:-Write a program that accepts a square matrix A and a positive integer m as arguments and returns A^m.

```python
import numpy as np

def matrix_power(A, m):
    # Checks if A is a square matrix
    rows, cols = A.shape
    if rows != cols:
        return "Error: Matrix must be square"

    # Checks if m is positive
    if m <= 0:
        return "Error: Exponent must be a positive integer"

    # Computes A^m
    return np.linalg.matrix_power(A, m)


# Main program
if __name__ == "__main__":

    A = np.array([[2, 1],
                  [1, 2]])
    m = 3

    result = matrix_power(A, m)
    print("A^m =\n", result)
```

Q4:-Write a program to count the highest occurring character & its occurrence count in an input string. Consider only alphabets. Ex: for “hippopotamus” as input string, the maximally occurring character is ‘p’ & occurrence count is 3.

```python
def highest_occurring_char(s):
    # it filters only alphabets
    filtered = [ch.lower() for ch in s if ch.isalpha()]

    if not filtered:
        return "No alphabets found"

    # it Counts occurrences
    freq = {}
    for ch in filtered:
        freq[ch] = freq.get(ch, 0) + 1

    # Finds the character with maximum frequency
    max_char = max(freq, key=freq.get)
    max_count = freq[max_char]

    return max_char, max_count


# Main program
if __name__ == "__main__":
    input_string = input("Enter a string: ")
    result = highest_occurring_char(input_string)

    if isinstance(result, str):
        print(result)
    else:
        char, count = result
        print(f"The maximally occurring character is '{char}' with occurrence count {count}")
```

Q5:-Generate a list of 25 random numbers between 1 and 10. Find the mean, median and mode for these numbers.

```python
import random
import statistics

def give_random_numbers():
    # it Generates 25 random integers between 1 and 10
    numbers = [random.randint(1, 10) for _ in range(25)]

    # Calculate mean, median, and mode
    mean_val = statistics.mean(numbers)
    median_val = statistics.median(numbers)
    mode_val = statistics.mode(numbers)

    return numbers, mean_val, median_val, mode_val


# Main program
if __name__ == "__main__":
    nums, mean_val, median_val, mode_val = give_random_numbers()
    print("Random Numbers:", nums)
    print("Mean:", mean_val)
    print("Median:", median_val)
    print("Mode:", mode_val)
```
